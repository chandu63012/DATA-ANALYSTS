```python
# Drop rows with NaN in CancellationCode
cancel_data = df[df['Cancelled'] == 1]['CancellationCode'].dropna()

# Count occurrences of each cancellation reason
cancel_counts = cancel_data.value_counts()

# Define labels (based on DOT codes)
labels = {
    'A': 'Carrier',
    'B': 'Weather',
    'C': 'NAS',
    'D': 'Security',
    'E':  'Late Aircraft',
}

# Map codes to labels
cancel_counts.index = cancel_counts.index.map(labels)

# Plot pie chart
plt.figure(figsize=(3,2))
plt.pie(cancel_counts, labels=cancel_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Flight Cancellations by Reason in 2008')
plt.show()
```

```python
# Group by Month and count flights
monthly_flights = df.groupby('Month').size()

# Plotting moths
plt.figure(figsize=(5,2))
sns.barplot(x=monthly_flights.index, y=monthly_flights.values, palette='viridis')
plt.title('Monthly Flight Counts in 2008')
plt.xlabel('Month')
plt.ylabel('Number of Flights')
plt.xticks(range(0,12), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.show()
```

```python
# Count flights by day of the week
week_flights = df['DayOfWeek'].value_counts().sort_index()

# Now plot
plt.figure(figsize=(5,2))
sns.barplot(x=week_flights.index, y=week_flights.values, palette='viridis')
plt.title('Weekly Flight Counts in 2008')
plt.xlabel('Day of Week(1=Mon,2=Tues,3=wens,4=Thues,5=Fri,6=Sat,7=Sun)')
plt.ylabel('Number of Flights')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample plot comparing Taxi Out and Taxi In
plt.figure(figsize=(8, 4))
plt.plot(df['TaxiOut'], label='Taxi Out', color='orange')
plt.plot(df['TaxiIn'], label='Taxi In', color='green')
plt.title('Taxi Out vs Taxi In Times')
plt.xlabel('Flight Index')
plt.ylabel('Time (minutes)')
plt.legend()
plt.grid(True)
plt.show()
```
